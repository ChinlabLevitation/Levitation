```python
import numpy as np
import matplotlib.pyplot as plt

kvals = np.array([0.05, 0.1, 0.15, 0.2, 0.3, 0.4, 0.6, 1, 2, 4, 6, 10, np.inf])
htvals = np.array([-0.007, -0.046, -0.115, -0.208, -0.412, -0.602, -0.903, -1.259, -1.600,
                   -1.782, -1.840, -1.884, -1.942])

knvals = np.array([0.09611, 0.1922, 0.2883, 0.3844, 0.5767, 0.7689, 1.153, 1.922, 3.845, 7.689, 11.53, 19.22, np.inf])
htvals = -htvals

def f_T(knudsenNumber):
    return (9*np.power(knudsenNumber, 3))/(1+4.4844*np.power(knudsenNumber, 2))/(1+knudsenNumber)
htfit = f_T(knvals)
def newf_T(knudsenNumber):
    ft = -.2906*knudsenNumber**2 + 1.354*knudsenNumber - .2706
    return ft
htfit2 = newf_T(knvals)

plt.figure()
plt.plot(knvals, htvals, 'o', label='Takata et. al. (1994)')
plt.plot(knvals, htfit, '-', label='Empirical fit 1')
plt.plot(knvals, htfit2 , '-', label='Empirical fit 2')
plt.xscale('log')

plt.xlabel('Knudsen number')
plt.ylabel('Theromophoretic coefficient')
plt.xlim(0 , 1)
plt.legend()
plt.show()
```
